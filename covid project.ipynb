```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url)
soup
```

```python
code = BeautifulSoup(soup.text, "lxml")
table_code = code.table
table_code
```

```python
tags = table_code.find_all('tr')
tags
```

```python
data = []
for tag in tags:
  y = tag.text.split('\n')[1:]
  if y[0] != "":
    data.append(y)
```

```python
import csv
file = open('covid_data.csv','w')
x = csv.writer(file)
x.writerows(data)
file.close()
```

```python
import pandas as pd
df = pd.read_csv('covid_data.csv',encoding = 'latin1')
df
```
